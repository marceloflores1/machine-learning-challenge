```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna().replace({'CONFIRMED': 0, 'CANDIDATE': 1, 'FALSE POSITIVE': 2})
df.head()
```

# Select your features (columns)

```python
dfCorr = df.corr()
dfCorr1 = dfCorr[['koi_disposition']].sort_values(by='koi_disposition', ascending=False)
dfCorr1.head(10)
```

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_ss','koi_fpflag_co','koi_fpflag_nt','koi_fpflag_ec','koi_teq','koi_depth','koi_model_snr']]
X = selected_features
X
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df[['koi_disposition']]
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data

from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
```

# Train the Model



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

target_names = ['CONFIRMED', 'CANDIDATE', 'FALSE POSITIVE']

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

model2 = KNeighborsClassifier(n_neighbors=7)
model2.fit(X_train_scaled, y_train)

print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")

predictions = model2.predict(X_test_scaled)
report = classification_report(predictions, y_test, target_names=target_names, output_dict=True)
report = pd.DataFrame(data=report)
print(report)
```
